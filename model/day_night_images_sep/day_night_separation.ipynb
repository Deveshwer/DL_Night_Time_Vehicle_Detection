```python
import json

# Open the file containing the JSON data
with open('filename.json') as f:
    # Load the JSON data from the file
    data = json.load(f)

# Print the data to verify it was read correctly

# print(data)

# move files from one folder to another

import shutil

# Define the path of the file to move
file_base_path = 'path/to/'

# Define the path of the destination folder
dest_folder_day = 'path/to/destination/day/'
dest_folder_night = 'path/to/destination/night/'

# Use the shutil.move() method to move the file to the destination folder
# shutil.move(file_path, dest_folder)


# search for daytime or nighttime value in timeofday key

for item in data:
    if item['timeofday'] == 'daytime':
        file_path = file_base_path + item['image']
        shutil.move(file_path, dest_folder_day)
    else if item['timeofday'] == 'nighttime':
        file_path = file_base_path + item['image']
        shutil.move(file_path, dest_folder_night)
    
```
